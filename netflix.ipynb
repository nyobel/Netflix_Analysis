## Visualization of data

Frequency table and graph

```python
count_1 = df['type'].value_counts()
persnt_1 = df['type'].value_counts(normalize=True)*100

frq_table_1 = pd.DataFrame({'frequency': count_1 , 'Percentage' : persnt_1 })

count_1.plot(kind='pie')

frq_table_1
```

```python
count_2 = df['cast'].value_counts()
persnt_2= df['cast'].value_counts(normalize=True)*100

frq_table_2= pd.DataFrame({'frequency': count_2, 'Percentage' : persnt_2})

count_2.plot(kind='kde')

frq_table_2
```

```python
count_3 = df['country'].value_counts()
persnt_3 = df['country'].value_counts(normalize=True)*100

frq_table_3 = pd.DataFrame({'frequency': count_3 , 'Percentage' : persnt_3 })

count_3.plot(kind='pie')

frq_table_3
```

```python
count_6 = df['rating'].value_counts()
persnt_6 = df['rating'].value_counts(normalize=True)*100
freq_table_6 = pd.DataFrame({'Frequancy':count_6 , 'Persenteg':persnt_6})

count_6.plot(kind='bar')

freq_table_6
```
